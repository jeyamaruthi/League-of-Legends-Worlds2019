## Teamwise Analysis

```python
## In the dataset, the position column holds for specfic roles as well as there is a seperate value to filter the 
## entire dataset with respect to teams. Hence:
teamDataFrame = data[data['position']== 'Team']
teamDataFrame.head()
```

```python
## For Player DataFrame:
playerDataFrame = data[data['position']!= 'Team']
playerDataFrame.head()
```

```python
teamAboveAverage =  teamDataFrame[teamDataFrame['gamelength'] >= teamDataFrame['gamelength'].mean()]
x = list(teamAboveAverage['team'].unique())
```

```python
teamDataFrame2 = teamDataFrame[teamDataFrame['result']==1]
teamDataFrame2.head()
```

### Team/side analysis

```python
groupedTeams = teamDataFrame.groupby(['team','side']).agg({"gamelength":'mean',"teamkills":'mean',
                                                           "teamdeaths":'mean',"fbtime":'mean',"teamdragkills":'mean'})
groupedTeams['teamkills'] = groupedTeams['teamkills'].apply(np.floor)
groupedTeams['teamdeaths'] = groupedTeams['teamdeaths'].apply(np.floor)
groupedTeams['teamdragkills'] = groupedTeams['teamdragkills'].apply(np.floor)
groupedTeams.head()
```

#### Team with low tempo & slow games

```python
overallTeam = teamDataFrame2.groupby(['team']).agg({"gamelength":'mean',"teamkills":'mean',
                                                           "teamdeaths":'mean',"fbtime":'mean',"teamdragkills":'mean'})
slowTeams = overallTeam[overallTeam['gamelength'] >= teamDataFrame['gamelength'].mean()]
slowTeamsLists = list(slowTeams.index)
slowTeams
```

### Fast paced teams

```python
overallTeam2 = teamDataFrame2.groupby(['team']).agg({"gamelength":'mean',"teamkills":'mean',
                                                           "teamdeaths":'mean',"fbtime":'mean',"teamdragkills":'mean'})
```

```python
fastTeams = overallTeam2[overallTeam2['gamelength'] < teamDataFrame['gamelength'].mean()]
fastTeams
```

```python
PlayerStatsDataFrame = data.groupby(['player']).agg({"k":'mean',"d":'mean', "a":'mean',"dmgtochamps":'mean',
                                                      "wards":'mean',"cspm":"mean","gdat10":"mean","gdat10":"mean",
                                                      "dmgtochampsperminute":"mean"})
```

```python
PlayerStatsDataFrame['KDA'] = (PlayerStatsDataFrame['k'] + PlayerStatsDataFrame['a'])/ PlayerStatsDataFrame['d']
PlayerStatsDataFrame
```

#### Top 10 players based on KDA

```python
Top10KDAplayers = PlayerStatsDataFrame.copy()
Top10KDAplayers= Top10KDAplayers.sort_values(by=['KDA'],ascending = False)
Top10KDAplayers = Top10KDAplayers[1:11]
Top10KDAplayers[['KDA']]
```

```python
Top10KDAplayers.plot(y=["KDA"], kind="bar")
```
